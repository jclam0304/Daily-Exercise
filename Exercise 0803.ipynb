```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,50)
```

```python
y = 1.2*x+0.8
```

```python
plt.scatter(x,y)
plt.plot(x,y,"r")
```

```python
y = 1.2*x+0.8+0.6*np.random.randn(50)
```

```python
plt.scatter(x,y)
plt.plot(x,1.2*x+0.8, "r")
```
